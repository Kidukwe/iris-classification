```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv("/Users/kirill/Jupyter/02_projects/Iris_Classification/data/iris_data.csv")
X = df.drop("target", axis=1)
y = df["target"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
    )
```

```python
clf1 = DecisionTreeClassifier(random_state=42)
clf1.fit(X_train, y_train)
y_pred1 = clf1.predict(X_test)
print("Model DecisionTreeClassifier: full features")
print("Accuracy:",accuracy_score(y_test, y_pred1))
print("Cofusion matrix:\n", confusion_matrix(y_test, y_pred1))
```

```python
plt.figure(figsize=(12,8))
plot_tree(
    clf1, 
    feature_names=X.columns,
    class_names=df["target"].unique().astype(str), #uniqe   ,      0,1,2,   ; astype(str)     
    filled=True #  
)
plt.show()
```
